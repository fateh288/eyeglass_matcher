```python
import pandas as pd
import numpy as np

csv_file = 'csv/lenskart_products_categorical.csv'

df = pd.read_csv(csv_file,
                 usecols=['product_id',
                          'parent_category',
                          'frame_shape'],
                 dtype={"product_id": np.int32,
                        "parent_category":"category",
                        "frame_shape":"category"})
for ind in df.index:
     print(df['product_id'][ind])
```

```python
df.insert(3,"encodings",None,True)
print(df)
```

```python
from keras.preprocessing import image
from tensorflow import keras

loaded = keras.models.load_model("saved_models_autoencoder/basic_attempt1_deep")

img_width = 256
img_height = 256

for ind in df.index:
     print(df['product_id'][ind])
     file_name1 = 'provided_dataset_train/train/'+ str(df['product_id'][ind]) + ".jpg"
     file_name2 = 'provided_dataset_validate/validate/'+ str(df['product_id'][ind]) + ".jpg"
     print(file_name1,file_name2)
     try:
        img = image.load_img(file_name1, target_size=(img_width, img_height))
     except:
         pass
     try:
         img = image.load_img(file_name2, target_size=(img_width, img_height))
     except:
         pass
     try:
         img = image.img_to_array(img)
         img = np.expand_dims(img, axis=0)
         print(img.shape)
         encoded_img = loaded.encoder(img).numpy()
         flattened = np.ravel(encoded_img)
         df['encodings'][ind] = flattened
     except:
         print(df['product_id'][ind]," not found or dimension mismatch error")

print(df.head())
```

```python
df.to_csv('csv/lenskart_products_categorical.csv')
```
